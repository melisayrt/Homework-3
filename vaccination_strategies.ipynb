```python
# Import required libraries
import networkx as nx
import random
import matplotlib.pyplot as plt
```

#### Question 1: Optimal Vaccination (Known Network)

```python


# Create or load your network (replace with actual network)
G = nx.karate_club_graph()  # Example network

# Find node with maximum degree
max_degree_node = max(G.degree(), key=lambda x: x[1])[0]
max_degree = G.degree[max_degree_node]

print(f"Optimal node to vaccinate: {max_degree_node} (Connections: {max_degree})")
```

```python
##"When the complete network structure is known, we should vaccinate the node with the highest degree (most connections). 
# This maximizes the number of immunized relationships by blocking transmission through all direct contacts."
```

### Question 2: Random Vaccination Simulation

```python
def random_vaccination_simulation(G, T=10000):
    total = 0
    nodes = list(G.nodes())
    for _ in range(T):
        node = random.choice(nodes)
        total += G.degree[node]
    return total / T

print("Expected immunized (random):", random_vaccination_simulation(G))
```

### Question 3: Indirect Random Vaccination

```python
def indirect_vaccination_simulation(G, T=10000):
    total = 0
    nodes = list(G.nodes())
    for _ in range(T):
        node = random.choice(nodes)
        neighbors = list(G.neighbors(node))
        if neighbors:
            vaccinated = random.choice(neighbors)
            total += G.degree[vaccinated]
    return total / T

print("Expected immunized (indirect):", indirect_vaccination_simulation(G))
```
